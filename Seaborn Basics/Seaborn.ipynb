# PYTHON - SEABORN

### importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

### Loading the dataset

```python
data = pd.read_csv('tips.csv')
data.head(5)
```

# Scatterplots

```python
sns.scatterplot(data=data, x='total_bill', y='tip')
plt.show()
```

```python
# Setting the style of plots - Darkgrid 
sns.set(style='darkgrid')
```

```python
sns.scatterplot(data=data, x='total_bill', y='tip')
plt.show()
```

```python
# Basic scatter plot using "relplot"
sns.relplot(data=data, x='total_bill', y='tip')
plt.show()
```

```python
# Setting Hue parameter to identify smoker category easily 
sns.scatterplot(data=data, x='total_bill', y='tip', hue='smoker')
plt.show()
```

```python
# Setting Hue and Style parameter to identify smoker and time category easily 
sns.scatterplot(data=data, x='total_bill', y='tip', hue='smoker', style='time')
plt.show()
```

```python
# Difference between the scatterplot and relplot
sns.relplot(data=data, x='total_bill', y='tip', hue='smoker', style='time')
plt.show()
```

```python
sns.relplot(data=data, x='total_bill', y='tip', hue='size')
plt.show()
```

```python
# Varyinig the size of the points using size parameter
sns.relplot(data=data, x='total_bill', y='tip', size='size')
plt.show()
```

```python
# Setting the limit of size 
sns.relplot(data=data, x='total_bill', y='tip', size='size', sizes=(15,200))
plt.show()
```

# Subplots

```python
# Create a figure and axes using subplots with 4 rows and 3 columns
fig, axes = plt.subplots(nrows=4, ncols=3, figsize=(15, 20))

# Flatten the axes array for easy indexing
axes = axes.flatten()

# Scatter plots with different parameters
sns.scatterplot(x="total_bill", y="tip", data=data, ax=axes[0])
sns.scatterplot(x="total_bill", y="tip", data=data, hue="smoker", ax=axes[1])
sns.scatterplot(x="total_bill", y="tip", data=data, hue="smoker", style="time", ax=axes[2])
sns.scatterplot(x="total_bill", y="tip", data=data, hue="size", palette="ch:r=-5,l=.75", ax=axes[3])
sns.scatterplot(x="total_bill", y="tip", data=data, size="size", ax=axes[4])
sns.scatterplot(x="total_bill", y="tip", data=data, size="size", sizes=(15, 200), ax=axes[5])
sns.scatterplot(x="total_bill", y="tip", data=data, ax=axes[6])
sns.scatterplot(x="total_bill", y="tip", data=data, hue="time", ax=axes[7])
sns.scatterplot(x="total_bill", y="tip", data=data, hue="smoker", style="time", ax=axes[8])
sns.scatterplot(x="total_bill", y="tip", data=data, hue="size", palette="ch:r=-5,l=.75", ax=axes[9])
sns.scatterplot(x="total_bill", y="tip", data=data, size="size", ax=axes[10])
sns.scatterplot(x="total_bill", y="tip", data=data, size="size", hue="smoker", ax=axes[11])

# Set titles for each plot
axes[0].set_title("Scatter Plot 1")
axes[1].set_title("Scatter Plot 2")
axes[2].set_title("Scatter Plot 3")
axes[3].set_title("Scatter Plot 4")
axes[4].set_title("Scatter Plot 5")
axes[5].set_title("Scatter Plot 6")
axes[6].set_title("Scatter Plot 7")
axes[7].set_title("Scatter Plot 8")
axes[8].set_title("Scatter Plot 9")
axes[9].set_title("Scatter Plot 10")
axes[10].set_title("Scatter Plot 11")
axes[11].set_title("Scatter Plot 12")

# Adjust the layout to avoid overlapping labels
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# Create a figure and axes using subplots with 3 rows and 4 columns
fig, axes = plt.subplots(3, 4,  figsize=(15,10))

sns.lineplot(data=data, x="total_bill", y="tip", ax=axes[0,0])
axes[0,0].set_title('Lineplot')

sns.lineplot(data=data, x="total_bill", y="tip", hue='smoker', ax=axes[0,1])
axes[0,1].set_title('Lineplot - Bill per tip per smokers')

sns.lineplot(data=data, x="total_bill", y="tip", style='sex', ax=axes[0,2])
axes[0,2].set_title('Lineplot - Bill per tip per sex')

sns.lineplot(data=data, x="total_bill", y="tip", hue='size', ax=axes[0,3])
axes[0,3].set_title('Lineplot - Bill per tip per size')

sns.lineplot(data=data, x="size", y="tip", ci=None, ax=axes[1,0])
axes[1,0].set_title('Lineplot - Tip per Size')

sns.lineplot(data=data, x="size", y="tip", ci="sd", ax=axes[1,1])
axes[1,1].set_title('Lineplot - Tip per Size with SD')

sns.lineplot(data=data, x="time", y="tip", ax=axes[1,2])
axes[1,2].set_title('Lineplot - Tip per Time')

sns.lineplot(data=data, x="smoker", y="tip", ax=axes[1,3])
axes[1,3].set_title('Lineplot - Tip per smoker')

sns.lineplot(data=data, x="day", y="tip", hue='sex', ci=None, ax=axes[2,0])
axes[2,0].set_title('Lineplot - Tip per day per sex')

sns.lineplot(data=data, x="day", y="tip", hue='smoker', ci=None, ax=axes[2,1])
axes[2,0].set_title('Lineplot - Tip per day per smoker')

sns.lineplot(data=data, x="size", y="tip", hue='smoker', ci=None, ax=axes[2,2])
axes[2,2].set_title('Lineplot - Tip per size per smoker')

sns.lineplot(data=data, x="size", y="tip", hue='sex', ci=None, ax=axes[2,3])
axes[2,3].set_title('Lineplot - Tip per size per sex')

# Adjust the spacing between subplots
plt.tight_layout()

# Show the plots
plt.show()
```

```python
fig, axes = plt.subplots(3, 4, figsize=(16,8))

# Box plots
sns.boxplot(data=data, x='day', y='total_bill', ax = axes[0, 0])
axes[0, 0].set_title('Boxplot-Total bills per day')

sns.boxplot(data=data, x='day', y='total_bill', hue='smoker', ax = axes[0, 1])
axes[0, 1].set_title('Boxplot-Total bills per day per smoker catgory')

data['weekend']=data['day'].isin(['Sat', 'Sun'])
sns.boxplot(data=data, x='weekend', y='total_bill', ax = axes[0, 2])
axes[0, 2].set_title('Boxplot-Total bills per weekend')

sns.boxplot(data=data, x='size', y='total_bill', ax = axes[0, 3])
axes[0, 3].set_title('Boxplot-Total bills per category')

#Violin plots
sns.violinplot(data=data, x='day', y='total_bill', ax = axes[1, 0])
axes[1, 0].set_title('Violinplot-Total bills per day')

sns.violinplot(data=data, x='day', y='total_bill', hue='weekend', dodge=False, bw=0.15, cut=0, ax = axes[1, 1])
axes[1, 1].set_title('Violinplot without dodge-Total bills per day')

sns.violinplot(data=data, x='day', y='total_bill', hue='sex', split=True, ax = axes[1, 2])
axes[1, 2].set_title('Violinplot with splitted sex-Total bills per day')

sns.violinplot(data=data, x='day', y='total_bill', hue='sex', inner='stick', kind='violin', split=True, ax = axes[1, 3])
axes[1, 3].set_title('Violinplot with inner sticks-Total bills per day')

# Boxen plots
sns.boxenplot(data=data, x='day', y='total_bill', ax = axes[2, 0])
axes[2, 0].set_title('Boxenplot -Total bills per day')

sns.boxenplot(data=data, x='day', y='total_bill', hue='sex', ax = axes[2, 1])
axes[2, 1].set_title('Boxenplot -Total bills per day per sex')

sns.boxenplot(data=data, x='size', y='total_bill', ax = axes[2, 2])
axes[2, 2].set_title('Boxenplot -Total bills per size')

sns.boxenplot(data=data, x='size', y='total_bill', hue='smoker', ax = axes[2, 3])
axes[2, 3].set_title('Boxenplot -Total bills per size per smoker')

plt.tight_layout()

plt.show()
```
